```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Описание датасета:

* __id__ - уникальный идентификатор дома
* __date__ -	дата продажи дома
* __price__	- стоимость продажи дома
* __bedrooms__ -	количество спален
* __bathrooms__	- количество ванных комнат (0.5 означает туалет без душа)
* __sqft_living__	- квадратные метры жилой площади
* __sqft_lot__	- квадратные метры общей площади участка
* __floors__	- количество этажей
* __waterfront__	- наличие набережной
* __view__	- оценка вида от 0 до 4, где 4 — лучший вид
* __condition__	- оценка состояния от 1 до 5, где 5 — лучшее состояние
* __grade__	- оценка качества конструкции и дизайна от 1 до 13, где 1–3 — плохое качество, 7 — среднее, 11–13 — высокое
* __sqft_above__	- квадратные метры дома, находящиеся над землёй
* __sqft_basement__	- квадратные метры дома, находящиеся под землёй
* __yr_built__	- год постройки дома
* __yr_renovated__	- год последнего ремонта дома
* __zipcode__	- почтовый индекс
* __lat__	- широта
* __long__	- долгота
* __sqft_living15__	- квадратные метры жилой площади у 15 ближайших соседей
* __sqft_lot15__	- квадратные метры общей площади участка у 15 ближайших соседей

```python
prices = df['price'].values
n_bins = 10
plt.figure(figsize=(8, 10))
plt.hist(prices, bins=n_bins)
plt.xlabel('Цена')
plt.ylabel('Количество')
plt.title('Распределение цен на недвижимость')
plt.show()
```

* Чем выше цена, тем меньше подобных объектов представлено на рынке.

* Большинство объектов недвижимости продаётся по цене примерно до 1 500 000. При этом максимальное количество объектов продаётся по цене примерно до 700 000–1 000 000.


*Распределение жилой квадратуры*

```python
plt.figure(figsize = (10,8))
sns.histplot(df['sqft_living'], bins = 15, color='green')
plt.xlabel('Квадратные метры жилой площади')
plt.ylabel('Количество квадратных футов')
plt.title('Распределение жилой квадратуры');
```

```python
plt.boxplot(df['sqft_living'])
plt.xlabel('Квадратные футы жилой площади')
plt.ylabel('Количество квадратных футов')
plt.title('Распределение жилой квадратуры');
plt.show()
```

> Большая часть предложений на рынке недвижимости находится в диапазоне от 500 до 6000 квадратных футов, при этом максимальное количество предложений представлено площадью в 2000 квадратных футов. Это позволяет сделать вывод, что наиболее востребованы среди покупателей и арендаторов помещения среднего размера.



*1.4. Изучите распределение года постройки*

```python
df['yr_built'].describe()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df['yr_built'], bins = range(1900, 2030, 10), color='pink')
ax.xaxis.set_major_locator(plt.IndexLocator(base = 10, offset = 0))
ax.grid()
for i in ax.patches:
    ax.annotate('%1.0f'%(i.get_height()), (i.get_x() + 2, i.get_height() + 70))
plt.title('Распределение недвижимости по году постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество домов')
plt.show()
```

* Наблюдается явный рост количества новой недвижимости после 1940, что свидетельствует о развитии строительной отрасли и увеличении спроса на жильё. Это может быть связано с улучшением экономической ситуации, ростом населения и изменением предпочтений потребителей.

* Пик строительства приходится на период после с 2000 по 2010 года, что может быть обусловлено различными факторами, включая государственную поддержку строительной отрасли, улучшение условий кредитования и повышение доступности жилья.

* Несмотря на общий рост, наблюдаются некоторые колебания в количестве новой недвижимости, построенной в разные годы. Это может быть связано с экономическими циклами, изменениями в государственной политике и другими внешними факторами.


# ЗАДАЧА 2

*2.1. Изучите распределение домов от наличия вида на набережную*

```python
data = df['waterfront'].value_counts()

plt.figure(figsize=(10, 8))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['Не имеют вид на набережную', 'Имеют вид на набережную']);
```



> График распределения домов в зависимости от наличия вида на набережную показывает, что подавляющее большинство домов (99,2%) не имеют такого вида. Это может быть связано с различными факторами, например, особенностями застройки или расположением набережной.


*2.2. Изучите распределение этажей домов*

```python
counts = df.groupby('floors').size()
colors = 'red', 'blue', 'green', 'yellow', 'purple'
counts.plot(kind='bar', color=colors)
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.title('Распределение этажей домов')
plt.show()
```

>  График распределения этажей домов показывает, что большинство домов имеют один или два этажа.



*2.3. Изучите распределение состояния домов*

```python
data = df['condition'].value_counts()
fig, ax = plt.subplots(figsize=(10, 8))
data.plot.barh(ax=ax, width=0.8, color='orange', align='center', label='Количество')
plt.title('Распределение состояния домов')
plt.xlabel('Количество домов')
plt.ylabel('Оценка состояния')
plt.show()
```

```python
data = df['condition'].value_counts()

plt.figure(figsize=(10, 8))
plt.pie(data, autopct='%1.1f%%', labels=data.index,);
```



> График распределения состояний домов показывает, что большинство домов находятся в хорошем и удовлетворительном состоянии, что является положительным фактором. Наличие менее 1% домов в плохом и ужасном состоянии указывает на эффективность мер по поддержанию и улучшению жилищных условий. Почти 8% домов в отличном состоянии также говорит о высоком качестве жилищного фонда.


# ЗАДАЧА 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 2)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0 # Заменяем значения корреляции меньше 0.3 на 0

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('Количество комнат')
plt.ylabel('Стоимость недвижимости')
plt.title('Распределение стоимости недвижимости по количеству комнат')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(data=df, x='sqft_living', y='price', color='green')
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Стоимость недвижимости')
plt.title('Зависимость стоимости от квадратуры жилой площади')
plt.show()
```

```python
plt.figure(figsize=(10, 10))

sns.boxplot(x=df['grade'], y=df['price'], whis=0.8)

plt.xlabel('Категория недвижимости')
plt.ylabel('Стоимость')
plt.title('Распределение цен в зависимости от категории недвижимости');
```

```python
#Влияние количества ванных комнат.
plt.figure(figsize=(12, 10))
sns.boxplot(data=df, x='bathrooms', y='price')
plt.xlabel('Количество ванных комнат')
plt.ylabel('Стоимость')
plt.title('Распределение стоимости недвижимости по количеству ванных комнат')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.lineplot(data=df, x='yr_built', y='price', color='red')
plt.xlabel('Год постройки')
plt.ylabel('Стоимость недвижимости')
plt.title('Распределение стоимости от года постройки')
plt.show()
```

Тепловая карта показала высокую корреляцию стоимости жилья с количеством спален, количеством ванных комнат, квадратными метрами жилой площади, категорией, квадратными метрами дома, находящимися над землёй, и квадратными метрами жилой площади у 15 ближайших соседей. Это указывает на то, что эти факторы оказывают значительное влияние на стоимость недвижимости.
Также была обнаружена умеренная корреляция стоимости с видом и квадратными метрами дома, находящимися под землёй. Это означает, что эти факторы также важны, но их влияние на стоимость меньше, чем у других факторов.
Широта не показала значительной корреляции со стоимостью жилья, что может указывать на то, что географическое расположение не является определяющим фактором в формировании стоимости недвижимости.

Значительным образом влияет на стоимость недвижимости ее категория(состояние и дизайн), соответственно чем выше категория, тем выше стоимость. Год постройки также влияет на стоимость: дома до 1940 года могут стоить подороже, ввиду возможной исторической ценности, или уникального расположекния в исторической части города, также рост цен замечен у домов новой постройки после 2000 года с колебаниями ввиду экономических проблем в серидине 2000х. Распределение стоимости недвижимости в зависимости от количества комнат показывает, что увеличение количества комнат (как жилых, так и ванных комнат)приводит к росту стоимости жилья.
